# R데이터마이닝 과제- 지원회사와 빅데이터/머신러닝/딥러닝과의 관련성

## 학번: 12121774
## 학과: 통계학과

### 지원회사: sk브로드밴드

저는 통계학과 전공이기 때문에 제 주관적으로 빅데이터를 판단하고 데이터 분석에 대해서 서술했습니다. 빅데이터가 대두되면서 통계학적 분석 말고도 전산학에서 많이 발전해온 방법들이 많이 생겨났습니다. 통계학에서 쓰는 분석 방법들은 가정들과 조건들이 까다로웠습니다. 그런데 데이터의 크기들이 커지면서 그 가정들과 조건들을 충분히 만족시킬 수 있는 방법들이 많이 발전했고 bias같은 것들은 충분히 줄일 수 있게 되었습니다. 요즘 기업의 데이터분석 팀에서는 통계학의 분석방법말고도 머신러닝, 딥러닝의 역량을 요구하고 있습니다. 아직 많이 배우지 못해서 완벽하게 알고 있지는 않습니다. 로지스틱 회귀,앙상블 기법(배깅, 부스팅)에 대한 이론과 개념들을 이우주 교수님께 조금 배웠을 뿐 아직 실습으로 데이터 분석을 해보지는 못했습니다. 그리고 라소, 릿지 회귀에 대한 개념도 수업시간에 조금 설명은 해주셨지만 아직 완벽하게 알지 못하고 loss function도 통계학과 전산학에서 쓰는 계산이 다르다고 들었습니다. 통계학과 전산학에서 보는 머신러닝 관점이 많이 다르다고 들었습니다. 그래서 이 수업은 통계학과 수업이 아니고 scsc의 수업, 즉 전산학에 가깝다고 생각하지만 교수님께서는 통계학을 전공하셨기 때문에 다양한 관점으로 배울 수 있을 거 같아서 수업을 꼭 듣고 싶었습니다. <br>
 요즘은 이미지 분석, 텍스트 분석, 영상처리 등 머신러닝 알고리즘들이 많이 쓰인다고 알고있습니다. 그러나 아직 접해보지 못했고 더 많은 공부가 필요하다고 생각합니다. 머신러닝과 딥러닝의 차이는 머신러닝은 컴퓨터에게 학습을 하도록 시키는(?) 것이지만 딥러닝은 그 학습조차 스스로 판단하여 학습을 합니다. 지도학습, 비지도학습, 강화학습 이렇게 3가지 학습방법이 있다고 합니다. 취업을 하고 싶긴 하지만 정부에서 지원해주는 AI관련 대학원에 진학도 고려하고 있습니다. 이런 분야에서 기본적인 요구 지식이 통계학이라고 들었습니다. 그래서 통계학 전공자들이 조금 유리하다는 글을 많이 봤습니다. 그런데 교수님이 통계학의 조건 가정등을 깨지 못하면 오히려 더 어려울 수도 있다고 하셔서 조금 걱정도 되었습니다. 그래도 통계학의 지식이 필요한 것은 확실하니 기회가 된다면 AI관련 분야로 깊이 있게 공부해보고 싶습니다. <br>
학부 수준에서는 아직 머신러닝, 딥러닝에 대한 강의가 많지 않은 것이 조금 아쉽습니다. 그래서 기업에서는 석사, 박사 학위를 많이 요구합니다. 학사 졸업으로 이쪽분야로 취업하려면 관련 경험을 증명할 수 있는 프로젝트가 있어야 가능성이 있기 때문에 이 수업으로 배운 것들을 가지고 프로젝트를 할 생각입니다. IT와 빅데이터 관련 업종은 프로젝트의 유무가 엄청 중요하다고 생각합니다. 서술보다는 허접한 제 생각을 많이 적었습니다. <br><br>
